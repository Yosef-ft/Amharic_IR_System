## Importing the csv we have saved earlier

```python
import pandas as pd
df = pd.read_csv('for_stat.csv')
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (7,7))
plt.plot(df.Rank, df.frequency)
plt.xscale('log')
plt.show()
```

## This shows why using quartiles doesn't work on calculating cutoffs
* Comparing to the graph the inter-quartile range is in the range of  $10^{4}$ but we can clearly see that after  $10^{3}$ the numbers have a frequency less than 5.

```python
Q1 = df['Rank'].quantile(0.25)
Q3 = df['Rank'].quantile(0.75)
IQR = Q3 - Q1
IQR
```

### Let's zoom in to view which part of words are significant

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (7,7))
plt.plot(df.Rank, df.frequency)
plt.xscale('log')
plt.ylim(0,10000)
plt.show()
```

## So we use a cut-off point between $10^{1}$ and  $10^{3}$

```python
plt.figure(figsize = (7,7))
plt.plot(df.Rank, df.frequency)
plt.xscale('log')
plt.axvline(1000, color ='red')
plt.axvline(2, color ='red')
plt.show()
```
