## Locating the files

```python
import os

dir_path = r"C:\Users\user\Documents\Data_science\IR real\pdfspider\downloads"

categories = os.listdir(dir_path)

my_dict = {} # my_dict contains the category name and what is inside it as a list(the books)
for category in categories:
    category_dir_path = os.path.join(dir_path, category)
    books = os.listdir(category_dir_path)
    my_dict[category] = books

to_be_extracted_dir = []
for key, value in my_dict.items(): # the value is the folder that contains the list of pdfs; the key is the category
    category = os.path.join(dir_path, key)
    for pdfs in value:
        book = os.path.join(category, pdfs)
        #pdf = os.listdir(book) # because the pdfs insidee the book folder are divided into sections the
        pdf = [file for file in os.listdir(book) if file.endswith('.pdf')]
        
        for individual_pdf in pdf:
            to_extract_pdf = os.path.join(book, individual_pdf)
            to_be_extracted_dir.append(to_extract_pdf)
```

## Defining a function to extract the PDFs to analyzable format

```python
import numpy as np
chunk_size = 25
chunks = np.array_split(to_be_extracted_dir, len(to_be_extracted_dir) // chunk_size)


from PyPDF2 import PdfReader
from pdf2image import convert_from_path
import pytesseract
from PIL import Image
import numpy as np

def extract_pdf(chunks_int): #chunks_int takes values as chunks[0]
    for dir_to_pdf in chunks_int:
        name = dir_to_pdf.split("\\")[-2]

        with open(dir_to_pdf, 'rb') as file:
            reader = PdfReader(file)
            no_of_pages = len(reader.pages)
            for i in range (no_of_pages):
                page = reader.pages[i]
                
                images = convert_from_path(dir_to_pdf, first_page=i+1, last_page=i+1, poppler_path=r'C:\Program Files\poppler-24.02.0\Library\bin')
                image = images[0]
                
                recognized_text = pytesseract.image_to_string(image, lang='amh')
                

                # Save the extracted text to a .txt file
                output_file = os.path.join(os.path.dirname(dir_to_pdf), f'{name}.txt')
                with open(output_file, 'a', encoding='utf-8') as f:
                    f.write(recognized_text)

                print("Text extracted and saved to", output_file)
```

### Running threads to execute the funtion faster. This is because the `pytesseract` takes time to convert the image to text file.

```python
from threading import Thread

def process_chunk(chunk):
    extract_pdf(chunk)

threads = [Thread(target=process_chunk, args=(chunk,)) for chunk in chunks]
for thread in threads:
    thread.start()
for thread in threads:
    thread.join()
```
